```python
# %% [markdown]
# # Î‘Î½Î¬Î»Ï…ÏƒÎ· ÎÎ¿Î¼Î¹ÎºÏÎ½ ÎšÎµÎ¹Î¼Î­Î½Ï‰Î½ - ÎœÎ­ÏÎ¿Ï‚ 1: Î•Î¾ÎµÏÎµÏÎ½Î·ÏƒÎ· Î”ÎµÎ´Î¿Î¼Î­Î½Ï‰Î½
# 
# **Î£Ï„ÏŒÏ‡Î¿Î¹:**
# 1. Î’Î±ÏƒÎ¹ÎºÎ® Î±Î½Î¬Î»Ï…ÏƒÎ· Î´Î¿Î¼Î®Ï‚ Î´ÎµÎ´Î¿Î¼Î­Î½Ï‰Î½
# 2. Î•Î¾ÎµÏÎµÏÎ½Î·ÏƒÎ· Î½Î¿Î¼Î¹ÎºÏÎ½ Î±Î½Î±Ï†Î¿ÏÏÎ½
# 3. Î£Ï„Î±Ï„Î¹ÏƒÏ„Î¹ÎºÎ® Î±Î½Î¬Î»Ï…ÏƒÎ· ÎºÎµÎ¹Î¼Î­Î½Ï‰Î½
# 4. ÎŸÏ€Ï„Î¹ÎºÎ¿Ï€Î¿Î¯Î·ÏƒÎ· ÎºÎ±Ï„Î±Î½Î¿Î¼ÏÎ½

# %% [markdown]
# ## 1. Î‘ÏÏ‡Î¹ÎºÎ¿Ï€Î¿Î¯Î·ÏƒÎ· ÎºÎ±Î¹ Î¦ÏŒÏÏ„Ï‰ÏƒÎ· Î’Î¹Î²Î»Î¹Î¿Î¸Î·ÎºÏÎ½

# %%
# Î’Î±ÏƒÎ¹ÎºÎ­Ï‚ Î²Î¹Î²Î»Î¹Î¿Î¸Î®ÎºÎµÏ‚
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import re
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Î¡Ï…Î¸Î¼Î¯ÏƒÎµÎ¹Ï‚ Î³ÏÎ±Ï†Î·Î¼Î¬Ï„Ï‰Î½
sns.set_theme(style="whitegrid")  # Î§ÏÎ®ÏƒÎ· Ï„Î¿Ï… seaborn styling
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12

# %% [markdown]
# ## 2. Î¦ÏŒÏÏ„Ï‰ÏƒÎ· ÎºÎ±Î¹ Î‘ÏÏ‡Î¹ÎºÎ® Î•Î¾Î­Ï„Î±ÏƒÎ· Î”ÎµÎ´Î¿Î¼Î­Î½Ï‰Î½

# %%
def load_and_validate_data(file_path: str) -> pd.DataFrame:
    """
    Î¦ÏŒÏÏ„Ï‰ÏƒÎ· ÎºÎ±Î¹ Î±ÏÏ‡Î¹ÎºÏŒÏ‚ Î­Î»ÎµÎ³Ï‡Î¿Ï‚ Î´ÎµÎ´Î¿Î¼Î­Î½Ï‰Î½
    
    Args:
        file_path (str): Î”Î¹Î±Î´ÏÎ¿Î¼Î® Î±ÏÏ‡ÎµÎ¯Î¿Ï…
        
    Returns:
        pd.DataFrame: ÎšÎ±Î¸Î±ÏÎ¹ÏƒÎ¼Î­Î½Î¿ dataset
    """
    try:
        # Î¦ÏŒÏÏ„Ï‰ÏƒÎ· Î´ÎµÎ´Î¿Î¼Î­Î½Ï‰Î½
        df = pd.read_csv(file_path)
        
        # Î’Î±ÏƒÎ¹ÎºÎ¿Î¯ Î­Î»ÎµÎ³Ï‡Î¿Î¹
        print("ğŸ“Š Î’Î±ÏƒÎ¹ÎºÎ¬ Î£Ï„Î±Ï„Î¹ÏƒÏ„Î¹ÎºÎ¬:")
        print(f"- Î£Ï…Î½Î¿Î»Î¹ÎºÎ­Ï‚ Ï…Ï€Î¿Î¸Î­ÏƒÎµÎ¹Ï‚: {len(df):,}")
        print(f"- Î£Ï„Î®Î»ÎµÏ‚: {', '.join(df.columns)}")
        print(f"\nğŸ” Î•Î»Î»Î¹Ï€Î® Î´ÎµÎ´Î¿Î¼Î­Î½Î±:\n{df.isnull().sum()}")
        
        # Î ÏÎ¿ÏƒÎ¸Î®ÎºÎ· Ï‡Î±ÏÎ±ÎºÏ„Î·ÏÎ¹ÏƒÏ„Î¹ÎºÏÎ½ ÎºÎµÎ¹Î¼Î­Î½Î¿Ï…
        df['text_length'] = df['case_text'].str.len()
        
        return df
    
    except FileNotFoundError:
        print(f"âŒ Î£Ï†Î¬Î»Î¼Î±: Î¤Î¿ Î±ÏÏ‡ÎµÎ¯Î¿ {file_path} Î´ÎµÎ½ Î²ÏÎ­Î¸Î·ÎºÎµ")
        raise
    except Exception as e:
        print(f"âŒ Î£Ï†Î¬Î»Î¼Î± ÎºÎ±Ï„Î¬ Ï„Î· Ï†ÏŒÏÏ„Ï‰ÏƒÎ·: {str(e)}")
        raise

# Î¦ÏŒÏÏ„Ï‰ÏƒÎ· Î´ÎµÎ´Î¿Î¼Î­Î½Ï‰Î½ Î¼Îµ Ï„Î¿ ÏƒÏ‰ÏƒÏ„ÏŒ path
df = load_and_validate_data("data/raw/legal_cases.csv")

# %% [markdown]
# ## 3. Î‘Î½Î¬Î»Ï…ÏƒÎ· ÎšÎ±Ï„Î±Î½Î¿Î¼Î®Ï‚ ÎšÎ±Ï„Î·Î³Î¿ÏÎ¹ÏÎ½

# %%
def analyze_case_distribution(df: pd.DataFrame) -> None:
    """
    Î‘Î½Î¬Î»Ï…ÏƒÎ· ÎºÎ±Î¹ Î¿Ï€Ï„Î¹ÎºÎ¿Ï€Î¿Î¯Î·ÏƒÎ· ÎºÎ±Ï„Î±Î½Î¿Î¼Î®Ï‚ Ï…Ï€Î¿Î¸Î­ÏƒÎµÏ‰Î½
    """
    try:
        # Î¥Ï€Î¿Î»Î¿Î³Î¹ÏƒÎ¼ÏŒÏ‚ ÎºÎ±Ï„Î±Î½Î¿Î¼Î®Ï‚
        case_dist = df['case_outcome'].value_counts()
        
        # Î”Î·Î¼Î¹Î¿Ï…ÏÎ³Î¯Î± Î³ÏÎ±Ï†Î®Î¼Î±Ï„Î¿Ï‚
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
        
        # Î¡Î±Î²Î´ÏŒÎ³ÏÎ±Î¼Î¼Î±
        sns.barplot(x=case_dist.values, y=case_dist.index, ax=ax1)
        ax1.set_title('ÎšÎ±Ï„Î±Î½Î¿Î¼Î® ÎšÎ±Ï„Î·Î³Î¿ÏÎ¹ÏÎ½ Î¥Ï€Î¿Î¸Î­ÏƒÎµÏ‰Î½')
        ax1.set_xlabel('Î Î»Î®Î¸Î¿Ï‚ Î¥Ï€Î¿Î¸Î­ÏƒÎµÏ‰Î½')
        
        # Î Î¯Ï„Î±
        ax2.pie(case_dist.values, labels=case_dist.index, autopct='%1.1f%%')
        ax2.set_title('Î Î¿ÏƒÎ¿ÏƒÏ„Î¹Î±Î¯Î± ÎšÎ±Ï„Î±Î½Î¿Î¼Î®')
        
        plt.tight_layout()
        plt.show()
        
        # Î•ÎºÏ„ÏÏ€Ï‰ÏƒÎ· ÏƒÏ„Î±Ï„Î¹ÏƒÏ„Î¹ÎºÏÎ½
        print("\nğŸ“ˆ Î£Ï„Î±Ï„Î¹ÏƒÏ„Î¹ÎºÎ¬ ÎšÎ±Ï„Î·Î³Î¿ÏÎ¹ÏÎ½:")
        for outcome, count in case_dist.items():
            percentage = (count/len(df)) * 100
            print(f"- {outcome}: {count:,} Ï…Ï€Î¿Î¸Î­ÏƒÎµÎ¹Ï‚ ({percentage:.1f}%)")
            
    except Exception as e:
        print(f"âŒ Î£Ï†Î¬Î»Î¼Î± ÎºÎ±Ï„Î¬ Ï„Î·Î½ Î±Î½Î¬Î»Ï…ÏƒÎ· ÎºÎ±Ï„Î±Î½Î¿Î¼Î®Ï‚: {str(e)}")
        raise

analyze_case_distribution(df)
```
